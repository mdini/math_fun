```python
import numpy as np
import scipy as sp
import math 
import matplotlib.pyplot as plt
```

```python
inf=int(1e4)
```

## Re(z)>1

```python
formula=plt.imread('zeta1.png')
plt.imshow(formula)
```

```python
def zeta1(z):
    f=lambda n: n**(-z)
    vf=np.vectorize(f)
    arr= vf(np.arange(1,inf,1))
    s= arr.sum()
    return s
```

## 1>Re(z)>0

```python
formula=plt.imread('zeta2.png')
plt.imshow(formula)
```

```python
def zeta2(z):
    f= lambda n: (-1)**(n-1) * n**(-z)
    vf=np.vectorize(f)
    arr= vf(np.arange(1,inf,1))
    s= arr.sum()
    s= s * 1/(1-2**(1-z))
    return s
```

## Re(z)<0

```python
formula=plt.imread('zeta3.png')
plt.imshow(formula)
```

```python
formula=plt.imread('tau.png')
plt.imshow(formula)
```

```python
def zeta3(z):
    tau_1_s= lambda x: x**(-z) * np.e**(-x) 
    tau= sp.integrate.quad(tau_1_s, 0, inf, complex_func=True)[0]
    s= 2**z * np.pi**(z-1) * np.sin(np.pi*z*.5)* tau * zeta1(1-z)
    return s
```

## Main function 

```python
def zeta(z):
    if z.real>1:
        return zeta1(z)
    elif z==1:
        return None
    elif z.real>0:
        return zeta2(z)
    else:
        return zeta3(z)
```
