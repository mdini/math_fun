```python
from sklearn import datasets
diabetes= datasets.load_diabetes()
x= diabetes['data']
y= diabetes['target']
```

```python
import tensorflow as tf 
import keras 
from keras import layers 
```

```python
#params 
nodes= [10, 16, 32, 64, 32, 16, 1]
activation= 'relu'
loss= 'mae'
metrics= ['mape']
learning_rate= .1
optimizer= keras.optimizers.Adam(learning_rate= learning_rate)
epochs= 100
batch_size= 200
validation_split= .1
```

```python
model= keras.Sequential([
    layers.Input((nodes[0],)),
    layers.Dense(nodes[1], activation=activation),
    layers.Dense(nodes[2], activation=activation),
    layers.Dense(nodes[3], activation=activation),
    layers.Dense(nodes[4], activation=activation),
    layers.Dense(nodes[5], activation=activation),
    layers.Dense(nodes[6], activation=activation),
])
```

```python
model.summary()
```

```python
model.compile(optimizer=optimizer, loss=loss, metrics=metrics)
```

```python
history=model.fit(x,y, batch_size=batch_size, 
                  epochs=epochs, validation_split=validation_split )
```

```python
import pandas as pd
pd.DataFrame(history.history).plot()
```
