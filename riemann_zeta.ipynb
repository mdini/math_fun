```python
import numpy as np
import math
```

```python
inf=int(1e3) #small value to avoid overflow error
```

```python
def sum1(n,z):
    f= lambda k: (-1)**k * math.comb(n,k) * (k+1)**(-z)
    vf= np.vectorize(f)
    arr=vf(range(0,n+1))
    s=arr.sum()
    return s
```

```python
def sum2(z):
    f= lambda n: 1/(2**(n+1)) * sum1(n,z)
    vf=np.vectorize(f)
    arr=vf(range(0,inf))
    s=arr.sum()
    return s
```

```python
zeta = lambda z: 1/(1-2**(1-z)) * sum2(z)
```

```python
zeta(-2.0) # -2 is  a trivial zero of zeta
```

```python
zeta(.5+14.13j) # .5+14.1
```
