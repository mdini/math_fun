```python
import numpy as np
```

```python
c= -.8 + 0.156j
```

```python
def iterations(z):
    i=0
    while (abs(z)<1e5) & (i<100):
        z= z**2 +c
        i+=1
    if i==100:
        return 0
    return i
```

```python
reals= np.arange(-2,2,.005)
imags= np.arange(-2,2,.005)
```

```python
julia= [[None for x in reals] for y in imags]
```

```python
for na,a in enumerate(reals):
    for nb,b in enumerate(imags):
        z= a+ 1j*b 
        julia[nb][na]=iterations(z)
```

```python
import seaborn as sns
```

```python
sns.heatmap(julia)
```
